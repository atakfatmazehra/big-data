```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

veriler = pd.read_csv("odev_tenis.csv")
print(veriler)

from sklearn import preprocessing
veriler2= veriler.apply(preprocessing.LabelEncoder().fit_transform) #topluca encode eder
print(veriler2)

durum= veriler2.iloc[:,:1] #birinci kısmı sadece encode etti
from sklearn import preprocessing
ohe=preprocessing.OneHotEncoder()
durum= ohe.fit_transform(durum).toarray()
print(durum)

havadurumu= pd.DataFrame(data=durum, index=range(14), columns=["o","r","s"]) #durumların ismini yazarak tabloda birleştirdik
sonveriler= pd.concat([havadurumu,veriler.iloc[:,1:3]], axis=1) #durumlar ve sıc,nem birleşti
sonveriler= pd.concat([sonveriler,veriler2.iloc[:,-2:]], axis=1) # tüm tablo ile ruzgar ve play birleşti
print(sonveriler)






from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(sonveriler.iloc[:,:-1], sonveriler.iloc[:,-1:], test_size=0.33, random_state=0)
#birinci bagımsız değişkenler(son kolon hariç) #ikinci bağımlı değişken ile tahmin edicek

from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)

#backward elimination
import statsmodels.api as sm
X= np.append(arr = np.ones((14,1)).astype(int),values=sonveriler.iloc[:,:-1], axis=1) 

X_l= sonveriler.iloc[:,[0,1,2,3,4,5]].values #tüm sutunlar alında
X_l = np.array(X_l, dtype=float)
model= sm.OLS(sonveriler.iloc[:,-1:],X_l).fit()
print(model.summary())

sonveriler= sonveriler.iloc[:,1:]

import statsmodels.api as sm
X= np.append(arr = np.ones((14,1)).astype(int),values=sonveriler.iloc[:,:-1], axis=1) 

X_l= sonveriler.iloc[:,[0,1,2,3,4]].values #tüm sutunlar alında
X_l = np.array(X_l, dtype=float)
model= sm.OLS(sonveriler.iloc[:,-1:],X_l).fit()
print(model.summary())

x_train= x_train.iloc[:,1:]
x_test= x_test.iloc[:,1:]
#print(x_train)
#print(x_test)

reg.fit(x_train,y_train)
y_pred= reg.predict(x_test)
print(y_pred)
```
