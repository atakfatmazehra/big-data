```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#DOSYAYI OKUMA
veriler = pd.read_csv("veriler.csv")
print(veriler)


ulke = veriler.iloc[:,0:1].values #birinci kolondakileri aldık
#print(ulke)
from sklearn import preprocessing
le= preprocessing.LabelEncoder()
ulke[:,0]= le.fit_transform(veriler.iloc[:,0])
ohe = preprocessing.OneHotEncoder() #ohe= 2li cevirir 0/1 yani
ulke = ohe.fit_transform(ulke).toarray() 
#print(ulke)




c = veriler.iloc[:,-1:].values #sondan birinci kolondakileri aldık
print(c)
from sklearn import preprocessing
le= preprocessing.LabelEncoder()
c[:,-1]= le.fit_transform(veriler.iloc[:,-1])
ohe = preprocessing.OneHotEncoder()
c = ohe.fit_transform(c).toarray() 
print(c)



from sklearn.impute import SimpleImputer
imputer= SimpleImputer(missing_values=np.nan, strategy='mean')
Yas= veriler.iloc[:,1:4].values
imputer=imputer.fit(Yas[:,1:4]) #fit=öğrenme metodudur
#sayısal verileri öğren
Yas[:,1:4]=imputer.transform(Yas[:,1:4])
#VERİLERİN BİRLEŞTİRİLMESİ
sonuc = pd.DataFrame(data=ulke, index = range(22), columns=["FR","TR","US"]) #22satırımız oldugu için
sonuc2= pd.DataFrame(data=Yas, index=range(22), columns=["boy","kilo","yas"])


cinsiyet= veriler.iloc[:,-1].values
sonuc3= pd.DataFrame(data=c[:,:1], index=range(22), columns=["cinsiyet"])
print(sonuc3) #cinsiyet kısmını tek sutünda 1/0 yaptık

tam_sonuc= pd.concat([sonuc,sonuc2], axis=1)
#axis=1 1.colonlar arasında eşleşme yapar
#concat=iki dataframe arasında işlem yapar, birleştirir

tam_sonuc2= pd.concat([tam_sonuc,sonuc3], axis=1)
print(tam_sonuc2)

#VERİ KÜMESİNİN EĞİTİLMESİ

from sklearn.model_selection import train_test_split
#verileri böleceğiz
#bağımsız(ülke, yas) ve bağımlı(hedef=cinsiyet) değişken
#yüzde33=test  yüzde67=train yani model eğitim için

x_train, x_test, y_train, y_test = train_test_split(tam_sonuc,sonuc3, test_size=0.33, random_state=0)

#x=test y=train
# yatayda 2  dikeyde 2 toplam 4 veri seti çıkar

#ÖZNİTELİK ÖLÇEKLENDİRME

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

#farklı dünyalardaki verileri aynı dünyaya çektik
#yani biri 100ler biri 10lar ama bunu standardscaler ile aynı boyuta indirgedik

X_train = sc.fit_transform(x_train)
X_test= sc.fit_transform(x_test)
#print(X_train)
#print(X_test)

from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)
boy= tam_sonuc.iloc[:,3:4].values
print(boy)
sol= tam_sonuc.iloc[:,:3]
sag= tam_sonuc.iloc[:,4:]
veri= pd.concat([sol,sag], axis=1)
print(veri)
#BOY SİLİNDİ TABLODAN


x_train, x_test, y_train, y_test = train_test_split(veri,boy,test_size=0.33, random_state=0)
r2= LinearRegression()
r2.fit(x_train,y_train)
y_pred = r2.predict(x_test)
#print(y_pred)
#print(y_test)

#GERİ ELEME, !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!11
import statsmodels.api as sm
X= np.append(arr = np.ones((22,1)).astype(int),values=veri, axis=1) #22ye 1lik bir tablo oluştur
X_l= veri.iloc[:,[0,1,2,3,4]].values #tüm sutunlar alında
X_l = np.array(X_l, dtype=float)
model= sm.OLS(boy,X_l).fit()
print(model.summary())

#5.sutün verimsiz çıktıgı için sildik. şimdi daha başarılı bir tablo !!!!!!!!!!!!!!!!!!!
X_l= veri.iloc[:,[0,1,2,3]].values
X_l = np.array(X_l, dtype=float)
model= sm.OLS(boy,X_l).fit()
print(model.summary())
```
